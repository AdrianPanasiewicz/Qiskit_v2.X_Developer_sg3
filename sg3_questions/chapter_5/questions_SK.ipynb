1. Suppose you build a GHZ state circuit for 3 qubits i.e. the state such that after the measurement it will collapse to either |000> or |111>. You choose a backend that has 156 qubits. Then, you transpile the circuit into the ISA one. How many qubits will have the ISA circuit?

A) 3 qubits

B) 153 qubits

C) 156 qubits

***Answer:***
<Details>
<br/>
C)  156 qubits
</Details>

2. You have a backend of 10 qubits. You want to run a circuit concerning only 1 qubit with an Estimator primitive. You transpiled the circuit and obtained the ISA observable which is `IIIXIIIIII`, `IIIZIIIIII`, with coeffs `[1.+0.j], [1+0.4.j]`. Why is this observable invalid?

A) The number of backend qubits and the original circuit one are incompatible with each other

B) Pauli strings can be composed only from the `I, X and Y` letters

C) Coefficients must be real numbers

D) None of the above

***Answer:***
<Details>
<br/>
C)  Coefficients must be real numbers
</Details>

3. You have the following observables given. You want to group them with `group_commuting` function, so you can compute them from a fingle measurement. However, you want to commute basing on the whole operator, not on a per-qubit basis - how can you do this?

A) `op.group_commuting(group_wise=True)`

B) `op.group_commuting(group_wise=False)`

C) `op.group_commuting(per_qubit=False)`

D) `op.group_commuting(op_as_whole=True)`


***Answer:***
<Details>
<br/>
A)  `op.group_commuting(group_wise=True)`
</Details>
