## Practice Questions

**1) If you do not specify the number of shots in `sampler.run(...)` or in the PUBs, what determines how many shots will be used?**

A)  The backend automatically selects a random number of shots

B)  The circuit determines the number of shots

C)  The value of `default_shots` in `SamplerOptions`

D)  The number of qubits in the circuit 

***Answer:***
<Details>
<br/>
C)  The value of `default_shots` in `SamplerOptions`
</Details>

---

**2. Which of the following is listed as a top-level attribute of `SamplerOptions`?**

A)  layout

B)  routing

C)  execution

D)  observables 

***Answer:***
<Details>
<br/>
C)  execution
</Details>

---

**3) Which of the following correctly updates a nested Sampler option after the Sampler has been created?**

A)  sampler.twirling.enable_gates = True

B)  sampler.options.twirling.enable_gates = True

C)  sampler.update(twirling.enable_gates=True)

D)  sampler.options.update(twirling=True) 

***Answer:***
<Details>
<br/>
B)  sampler.options.twirling.enable_gates = True
</Details>
