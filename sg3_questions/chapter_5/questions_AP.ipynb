**1. Which error suppression and error mitigation options are available for the Sampler primitive? Select all that apply.**

A) Zero-noise estimation

B) Pauli Twirling

C) Dynamical decoupling

D) Twirling Readout Error Extinction (T-REx)



***Answer:***
<Details>
<br/>
B) Pauli Twirling and C) Dynamical decoupling

</Details>

**2. Which code snippet correctly submits a quantum circuit `qc` for sampling probabilities on a specified `backend`?**


A)
```
sampler = Sampler(backend=backend)
job = sampler.run([qc])
```

B)
```
job = backend.run(qc, mode="sampler")
```

C)
```
sampler = Sampler(backend=backend)
job = sampler.run(qc)
```

D)
```
job = backend.run([qc], mode="sampler")
```

***Answer:***
<Details>
<br/>
A)
```
sampler = Sampler(backend=backend)
job = sampler.run([qc])
```
</Details>

B)
```
sampler.twirling.enable_gates = True
sampler.dynamical_decoupling.enable = True
```

C)
```
sampler.options.twirling.enable_gates = True
sampler.options.dynamical_decoupling.enable = True
```

D)
```
sampler.options.twirling = True
sampler.options.dynamical_decoupling = True
```

***Answer:***
<Details>

```
sampler.options.twirling.enable_gates = True
sampler.options.dynamical_decoupling.enable = True
```
