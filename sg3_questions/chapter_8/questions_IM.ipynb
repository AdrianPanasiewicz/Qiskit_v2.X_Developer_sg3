## Practice Questions

**1) What is the main difference between OpenQASM 2 and OpenQASM 3?**

A)  OpenQASM 3 does not support quantum gates

B)  OpenQASM 3 adds classical control flow and variables

C)  OpenQASM 2 supports hardware execution, OpenQASM 3 does not

D)  OpenQASM 3 only works with simulators 

***Answer:***
<Details>
<br/>
B)  OpenQASM 3 adds classical control flow and variables
</Details>

---

**2. Which function is used to convert a Qiskit `QuantumCircuit` into an OpenQASM 3 string?**

A)  qasm2.dumps(qc)

B)  qasm3.dumps(qc)

C)  qc.to_qasm3()

D)  export_qasm3(qc) 

***Answer:***
<Details>
<br/>
B)  qasm3.dumps(qc)
</Details>

---

**3) What is the purpose of the `bit` type in OpenQASM 3?**

A)  To represent a quantum bit

B)  To represent a classical bit

C)  To store floating-point values

D)  To define gate parameters 

***Answer:***
<Details>
<br/>
B)  To represent a classical bit
</Details>
