**1. Why will the following code fail to export the quantum circuit `qc`?**

```python
with open("qasm_file.txt", "w") as f:
    dumps(qc, f)
```

A) QASM files can only be exported to `.json` files.

B) An `Exporter` instance must be created first, from which the `dumps` method should be called.

C) The syntax is incorrect, it should be `f.dumps(qc)`.

D) The syntax is wrong, `dump(qc, f)` should be used to export to a file.

***Answer:***
<Details>
<br/>
D) The syntax is wrong, `dump(qc, f)` should be used to export to a file.
</Details>
