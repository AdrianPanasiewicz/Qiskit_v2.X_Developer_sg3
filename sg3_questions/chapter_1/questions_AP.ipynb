## **Question Formats**

Questions should follow one of three primary structures used in the exam:

1.	Code Output/Behavior: Present a short code snippet and ask what the resulting output, circuit state, or matrix representation will be. Ensure the code includes minor syntax nuances or logical tricks, as the exam often tests attention to detail (e.g., Qiskit's little-endian notation or the difference between instantiating a gate versus applying it).
2.	Code Completion: Provide a scenario or an incomplete code block and ask which option correctly completes the task. For example: "Which code fragment will produce a maximally entangled Bell state?" Include plausible distractors that might result in errors or incorrect circuit dimensions.
3.	Conceptual/Theoretical: Test the underlying mechanics of Qiskit components without code. For instance, ask about the dimensions of a specific operator, the attributes of a Pauli class, or the default phase applied by a specific generalized gate.

## **Distractor Design**

The distractors (incorrect options) must be carefully crafted to represent common misconceptions.

When writing options, designer should include:

1. A syntax error disguised as a valid method (e.g., using .cnot() instead of .cx() if testing older versus newer syntax).
2. A mathematically correct answer that is formatted incorrectly according to Qiskit standards (e.g., reversing the qubit order).
3. A partially correct answer that misses a key constraint mentioned in the prompt (e.g., asking for a circuit with 4 qubits and 4 classical bits, but providing QuantumCircuit(4) as an option).

## **Rule for creating questions**

Each person should create 3 questions from their assigned topic. Questions should follow these rules:

1. There should be one question for each question format (Code Output/Behavior, Code Completion, Conceptual/Theoretical).
2. Each question should be broad enough to expose weak understanding, but specific enough that there is one best answer.
3. Each person should also explain why the wrong options are wrong, because that is where the knowledge gaps usually appear.
