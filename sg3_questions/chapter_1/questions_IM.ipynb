## Practice Questions

**1) Which statement about the `Operator` class in Qiskit is true?**

A)  It stores operators as sparse Pauli strings

B)  It stores operators as dense matrices

C)  It can only represent single-qubit gates

D)  It is mainly used to store measurement counts 

***Answer:***
<Details>
<br/>
B)  It stores operators as dense matrices
</Details>

---

**2. Which operation is not supported when converting a `QuantumCircuit` to an `Operator`?**

A)  h gate

B)  cx gate

C)  measure operation

D)  x gate 

***Answer:***
<Details>
<br/>
C)  measure operation
</Details>

---

**3. Which function is used in Qiskit to compute the fidelity of two quantum channels?**

A)  channel_fidelity(opA, opB)

B)  process_fidelity(opA, opB)

C)  fidelity(opA, opB)

D)  state_fidelity(opA, opB) 

***Answer:***
<Details>
<br/>
B)  process_fidelity(opA, opB)
</Details>
