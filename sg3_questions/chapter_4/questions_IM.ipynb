## Practice Questions

**1) Given an Estimator instance and a quantum circuit `qc` with observable `H`, how do you correctly submit a job?**

A)  job = estimator.run(qc, H)

B)  job = estimator.run([(qc, H)])

C)  job = estimator.execute(qc, H)

D)  job = qc.run(estimator, H) 

***Answer:***
<Details>
<br/>
B)  job = estimator.run([(qc, H)])
</Details>

**2. What is the main advantage of using session mode when running multiple quantum jobs?**

A)  It allows execution without a backend

B)  It guarantees zero noise

C)  It reduces latency between consecutive jobs

D)  It skips transpilation 

***Answer:***
<Details>

</Details>
