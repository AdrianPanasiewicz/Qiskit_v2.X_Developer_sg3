**1.When working with variational quantum circuits that require consecutive executions where each circuit depends on the output of the previous run, which execution method is most appropriate?**

A) Job mode

B) Batch mode

C) Session mode



***Answer:***
<Details>
<br/>
C) Session mode

**2. Given a `sampler` instance and a quantum circuit `qc`, how would you submit a job for execution?**


A) `job = sampler.run([qc])`

B) `job = sampler.execute(qc)`

C) `job = sampler.run(qc)`

D) `job = qc.run(sampler)`

***Answer:***
<Details>
<br/>
A) `job = sampler.run([qc])`
</Details>
