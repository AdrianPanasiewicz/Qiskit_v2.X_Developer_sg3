## Practice Questions

**1) If neither a PUB nor `run()` specifies precision, what value is used by the Estimator by default?**

A)  0.1

B)  0.015625

C)  1 / 4096

D)  4096 

***Answer:***
<Details>
<br/>
B)  0.015625
</Details>

---

**2. What happens if `default_shots` is set in `EstimatorOptions`?**

A)  It is ignored unless `resilience_level` is also set

B)  It overrides `default_precision`

C)  It disables twirling automatically

D)  It can only be used on simulators 

***Answer:***
<Details>
<br/>
B)  It overrides `default_precision`
</Details>

---

**3) Which of the following correctly sets the default precision when creating an Estimator?**

A)  estimator = Estimator(default_precision=0.01)

B)  estimator = Estimator(options={"default_precision": 0.01})

C)  estimator = Estimator(); estimator.default_precision = 0.01

D)  estimator = Estimator(); estimator.options.default_precision = 0.01 

***Answer:***
<Details>
<br/>
D)  estimator = Estimator(); estimator.options.default_precision = 0.01
</Details>
