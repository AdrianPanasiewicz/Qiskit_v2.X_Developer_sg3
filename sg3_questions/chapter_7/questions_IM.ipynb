## Practice Questions

**1) Which method is commonly used to continuously monitor the status of a running job?**

A)  job.result()

B)  job.status()

C)  job.monitor()

D)  job.wait() 

***Answer:***
<Details>
<br/>
B)  job.status()
</Details>

---

**2. Why is it important to store the `job_id` of a submitted job?**

A)  It is required to run the job

B)  It allows you to retrieve the job and its results later

C)  It determines the backend used

D)  It speeds up execution 

***Answer:***
<Details>
<br/>
B)  It allows you to retrieve the job and its results later
</Details>

---

**3) Which of the following correctly retrieves a previously submitted job using its ID?**

A)  job = service.get(job_id)

B)  job = service.job(job_id)

C)  job = job.load(job_id)

D)  job = backend.job(job_id) 

***Answer:***
<Details>
<br/>
B)  job = service.job(job_id)
</Details>
