## Practice Questions

**1. Given an `EstimatorResult` object `results`, how do you retrieve the expectation values?**

A) `evs = results.data.evs`

B) `evs = results[0].data.evs`

C) `evs = results.evs`

D) `evs = results[0].evs`



***Answer:***
<Details>
<br/>
B) `evs = results[0].data.evs`

</Details>

**2. Given a `service` instance and a `job_id`, how do you retrieve the corresponding job?**


A) `retrieved_job = service.job(job_id)`

B) `retrieved_job = service.from_id(job_id)`

C) `retrieved_job = service.jobs[job_id]`

D) `retrieved_job = service.find_job(job_id)`

***Answer:***
<Details>
<br/>
A) `retrieved_job = service.job(job_id)`
</Details>

**3. What is the behavior of `job.result()` if the job has not yet finished executing when this line is reached?**

```
job = sampler.run([(isa_circuit, params)])
...
results = job.result()
```

A) An error is raised, but the job continues running in the background.

B) An error is raised and the job is aborted.

C) `results` is assigned `None` until the job completes.

D) `job.result()` blocks execution until the job finishes and the result is available.

***Answer:***
<Details>
<br/>
D) `job.result()` blocks execution until the job finishes and the result is available.
</Details>
