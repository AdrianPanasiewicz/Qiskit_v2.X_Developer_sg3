1. You take a look at the transpiled ISA circuit. You see it has no consecutive 1-qubit gates that are the inverses of each other, but you can see the number of qubits are the same as in the original quantum circuit. Furthermore, you see that on one of the qubits there is a number of rotations around the Y axis that could be summed up to one rotation by `π/2` radians. What would you do to fix both of those issues?


A) Switch from optimization level 0 to optimization level 1

B) Switch from optimization level 1 to optimization level 2

C) Switch from optimization level 3 to optimization level 2

***Answer:***
<Details>
<br/>
B) Switch from optimization level 1 to optimization level 2

**Explanation:** The lack of no consecutive 1-qubit gates being the inverses of each other indicates we have the optimization level of at least 1. The commutative cancellation and mapping to physical qubits from virtual one can occur either at optimization level 2 or 3
</Details>

A) ```
    parameters=[]
for i in range (k):
    newParam = Parameter(f"hi{i}")

    ```

B) ```
    import numpy as np

    parameters = ParameterVector("hi", k)
