## Practice Questions

**1) Which method measures only the qubits that have operations applied to them?**

A)  measure

B)  measure_all

C)  measure_active

D)  measure_used 

***Answer:***
<Details>
<br/>
C)  measure_active
</Details>

**2. Which Qiskit construct is used to conditionally apply operations based on the value of a classical bit or register?**

A)  with circuit.if_test(...)

B)  with circuit.while_loop(...)

C)  circuit.control_if(...)

D)  circuit.branch(...) 

***Answer:***
<Details>
<br/>
A)  with circuit.if_test(...)
</Details>

**3) What is the main purpose of the layout stage in Qiskit transpilation?**

A)  To decompose gates into the backend’s basis gates

B)  To assign virtual qubits in the circuit to physical qubits on the backend

C)  To insert SWAP gates for connectivity

D)  To remove redundant gates 

***Answer:***
<Details>
<br/>
B)  To assign virtual qubits in the circuit to physical qubits on the backend
</Details>
