**2. Given the image and the code below, how to finish the code to get the same circuit as the image?**

![HK_c2_2.png](q_photos/HK_c2_2.png)

```
qubits = QuantumRegister(2)
clbits = ClassicalRegister(2)
circuit = QuantumCircuit(qubits, clbits)
(q0, q1) = qubits
(c0, c1) = clbits

circuit.h(q0)
circuit.measure(q0, c0)
# [put the code here]
circuit.measure(q1, c1)

circuit.draw("mpl")

```


A)
```
with circuit.if_test((c0, 0)) as else_:
    circuit.h(q1)
with else_:
    circuit.x(q1)
```

B)
```
with circuit.if_test((c0, 1)) as else_:
    circuit.x(q1)
with else_:
    circuit.h(q1)
```

C)
```
with circuit.if_test((c0, 0)) as else_:
    circuit.x(q1)
with else_:
    circuit.h(q1)
```

D)
```
with circuit.if_test((c0, 1)) as else_:
    circuit.h(q1)
with else_:
    circuit.x(q1)
```

***Answer:***
<Details>
<br/>
C)
```
with circuit.if_test((c0, 0)) as else_:
    circuit.x(q1)
with else_:
    circuit.h(q1)
```
</Details>
